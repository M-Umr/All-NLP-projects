```
from tqdm import tqdm
import random
#!pip install gensim
#!pip install modules
import modules
from gensim.models import Word2Vec
```

```
data = pd.read_csv("dataset.csv")
data.head()
```

```
data['id'] = data['id'].astype(str)
data['model'] = data['model'].astype(str) 
```

```
stockid = data["id"].unique().tolist()
print(len(stockid))
random.shuffle(stockid)
stock_train = [stockid[i] for i in range(round(0.9*len(stockid)))]
train_stock = data[data['id'].isin(stock_train)]
validation_stock = data[~data['id'].isin(stock_train)]
```

```

# list to capture purchase history of the customers
purchases_train = []

# populate the list with the product codes
for i in tqdm(stock_train):
    temp = train_stock[train_stock["id"] == i]["model"].tolist()
    purchases_train.append(temp)
```

```
# list to capture purchase history of the customers
purchases_val = []

# populate the list with the product codes
for i in tqdm(validation_stock['id'].unique()):
    temp = validation_stock[validation_stock["id"] == i]["model"].tolist()
    purchases_val.append(temp)
```

```
model = gensim.models.Word2Vec(window = 10, sg = 1, hs = 0,negative = 10,alpha=0.03, min_alpha=0.0007,seed = 14)
#v = model.build_vocab(purchases_train,update=False)
#print(v)
```

```
model.train(purchases_train, total_examples = model.corpus_count,epochs=10, report_delay=1)
```
